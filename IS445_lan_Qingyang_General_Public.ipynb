```python
data_url = 'https://raw.githubusercontent.com/qlannn/cvs/main/all_seasons.csv'
df = pd.read_csv(data_url, encoding='latin1')
```

```python
df
```

For peers, those who like basketball analysis.

```python
rodman_df = df[df['player_name'] == 'Dennis Rodman']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(rodman_df['reb'], rodman_df['season'])
ax.set_ylabel('Season')
ax.set_xlabel('Rebounds')
ax.set_title('Dennis Rodman Rebounds performance')
plt.xticks(rotation=150)
plt.show()
```

From this visual example we can see the statistical performance of the players during the season, whether it is points, rebounds, etc. Can be changed to scores to form new example studies.

```python
import matplotlib.pyplot as plt

data = df[['Healthy life expectancy', 'Score']]

plt.scatter(data['Healthy life expectancy'], data['Score'], alpha=0.5)
plt.title('Relationship between Healthy life expectancy and Score')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Score')
plt.show()
```

This data visualization example illustrates the impact of Healthy life expectancy on happiness scores

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# to use interactivity
import ipywidgets
```

```python
data_url = 'https://raw.githubusercontent.com/qlannn/cvs/main/all_seasons.csv'
df = pd.read_csv(data_url, encoding='latin1')
```

```python
df
```

```python
player_name = 'Markelle Fultz'
season = '2017-18'
columns = ['team_abbreviation', 'pts', 'reb', 'ast','country','college','draft_round','player_height','draft_year','net_rating',
          'oreb_pct', 'dreb_pct', 'usg_pct', 'ts_pct', 'ast_pct']

fultz_2018 = df.loc[(df['player_name'] == player_name) & (df['season'] == season), columns]

print(fultz_2018)
```

```python

rodmans_rebs = df[df['player_name'] == 'Dennis Rodman']['reb'].sum()
curetons_rebs = df[df['player_name'] == 'Dwayne Schintzius']['reb'].sum()

total_rebs = [rodmans_rebs, curetons_rebs]

labels = ['Dennis Rodman', 'Dwayne Schintzius']
colors = ['yellow','pink']
explode = (0.2, 0) 

plt.pie(total_rebs, colors=colors, labels=labels, autopct='%1.1f%%', startangle=120, pctdistance=0.95, explode=explode)
plt.axis('equal')  
plt.title('Rodman vs Dwayne ') 
plt.show()
```

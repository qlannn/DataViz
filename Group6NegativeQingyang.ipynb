```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# to use interactivity
import ipywidgets
```

```python
data_url = 'https://raw.githubusercontent.com/qlannn/cvs/main/store_data.csv'
df = pd.read_csv(data_url, encoding='latin1')
```

```python
df
```

## First we need to show the data, and now you can see it clearly, I'm going to give you some visual examples of negative data

```python
orders_by_state = df.groupby('State')['Order ID'].count().sort_values()
orders_by_state.plot(kind='line')
plt.title('Order Count by State (Lowest to Highest)')
plt.xlabel('State')
plt.ylabel('Order Count', fontsize=7)
plt.show()
```

A negative example of a gradual response to the number of orders per state. You can see that some of the more remote states have very low orders.

```python
# Group the data by year and calculate the total sales
sales_by_year = df.groupby('Region')['Sales'].sum().sort_index()

# Create a line plot of sales by year
plt.plot(sales_by_year.index, sales_by_year.values, color='pink')

plt.title('Total Sales by Year')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

Simple example of data visualization, based on sales in different regions. You can see that there are big differences in different regions. Is a negative example.
